Feature Relationships and Ask the Qestions

```python
df_cut['price_per_m2']=df_cut['price']/df_cut['squareMeters']
```

```python
#nocorrelaction
sns.scatterplot(data=df_cut, x='NewBuildYear', y='price_per_m2')
ax = plt.gca()  
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: f'{y:,.0f}'))
```

```python
sns.scatterplot(data=df_cut, x='centreDistance', y='price_per_m2')
ax = plt.gca()  
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: f'{y:,.0f}'))
```

```python
sns.scatterplot(data=df_cut, x='squareMeters', y='price_per_m2')
ax = plt.gca()  
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: f'{y:,.0f}'))
```

```python
df_cut.columns
```

```python
#which cities have the most offers
top_cities = df_cut.groupby(['city']).count()['id'].sort_values(ascending=False).head(10)
top_plot = top_cities.plot(kind='barh', title='Top 10 cities with the most offers')
top_plot.set_xlabel('Amount of offers')
top_plot.set_ylabel('City')
```

```python
#which cities have the most expensive real estate

the_most_expensive_cities = df_cut.groupby('city')['price_per_m2'].mean().round().astype('int').sort_values(ascending=False).head(10)
ax = the_most_expensive_cities.plot(kind='barh', title='Top 10 cities by average housing price for m2')
ax.set_xlabel('Price')
ax.set_ylabel('City')
```
